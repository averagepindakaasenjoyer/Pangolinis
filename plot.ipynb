```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- Extracted Data ---
epochs = list(range(1, 11))

train_loss = [2.5132, 1.9006, 1.6481, 1.4963, 1.3924, 1.2756, 1.1399, 1.0367, 0.8938, 0.7573]
train_acc = [0.1817, 0.3643, 0.4398, 0.4949, 0.5200, 0.5632, 0.6190, 0.6673, 0.7225, 0.7749]
train_f1 = [0.1756, 0.3490, 0.4296, 0.4866, 0.5139, 0.5587, 0.6147, 0.6650, 0.7210, 0.7738]

val_loss = [2.0892, 1.7829, 1.6685, 1.6112, 1.5854, 1.6066, 1.6469, 1.5758, 1.6898, 1.6813]
val_acc = [0.3281, 0.3993, 0.4183, 0.4475, 0.4610, 0.4624, 0.4610, 0.4583, 0.4671, 0.4651]
val_f1 = [0.3100, 0.3917, 0.4091, 0.4388, 0.4478, 0.4562, 0.4513, 0.4509, 0.4648, 0.4628]

# Set a style for better aesthetics
sns.set_style("whitegrid")

# Create plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5)) # 1 row, 3 columns for Loss, Acc, F1

# Plot Loss
axes[0].plot(epochs, train_loss, label='Train Loss', marker='o', markersize=4, linestyle='-')
axes[0].plot(epochs, val_loss, label='Validation Loss', marker='o', markersize=4, linestyle='--')
axes[0].set_title('Training & Validation Loss per Epoch')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].legend()
axes[0].grid(True, linestyle=':', alpha=0.7)

# Plot Accuracy
axes[1].plot(epochs, train_acc, label='Train Accuracy', marker='o', markersize=4, linestyle='-')
axes[1].plot(epochs, val_acc, label='Validation Accuracy', marker='o', markersize=4, linestyle='--')
axes[1].set_title('Training & Validation Accuracy per Epoch')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy')
axes[1].legend()
axes[1].grid(True, linestyle=':', alpha=0.7)

# Plot F1 Score
axes[2].plot(epochs, train_f1, label='Train F1-Score (Weighted)', marker='o', markersize=4, linestyle='-')
axes[2].plot(epochs, val_f1, label='Validation F1-Score (Weighted)', marker='o', markersize=4, linestyle='--')
axes[2].set_title('Training & Validation F1-Score per Epoch')
axes[2].set_xlabel('Epoch')
axes[2].set_ylabel('F1-Score')
axes[2].legend()
axes[2].grid(True, linestyle=':', alpha=0.7)

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```
